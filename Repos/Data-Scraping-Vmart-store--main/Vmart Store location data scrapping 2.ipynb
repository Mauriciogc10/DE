```python
# importing libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# for coordinate extraction from link
import urllib.parse
```

```python
# Headers for request
HEADERS = ({ "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36", 'Accept-Language': 'en-US, en;q=0.5'})
```

```python
# list for required info
store_name =[]
address = []
phone_no = []
timing = []
links =[]
latitude =[]
longitude = []
```

```python
# for vmart in bihar no. of page for result 1 to 3
for i in range(1,4):
    url = "https://stores.vmartretail.com/location/bihar?page="+str(i)
    # using request to call link url
    Webpage = requests.get(url, headers =HEADERS)
    # using soup 
    soup = BeautifulSoup(Webpage.text, 'html.parser')
    # extracting data from ,limit section for reducing chance of geting extract data
    box = soup.find('div', class_ ='card-deck mt-3')
    # for etracting store name
    store = box.find_all('a', class_ ='col-store-name txt-color-3 txt-hover-1')
    # Extract the name from each div tag
    for i in store:
        s = i.text.strip()
        store_name.append(s)
        
    # extract address
    add = box.find_all("span", class_ = "store-address-info idx-info-card-str-add-info")
    for i in add:
        a = i.text.strip()
        address.append(a)
        
    # extract phone
    phone = box.find_all("a", class_ ="txt-hover-1 txt-color-1")
    for i in phone:
        ph = i.text.strip()
        phone_no.append(ph)
        
    # etract time
    time = box.find_all("span", class_="store-time-info idx-info-card-str-time-info")
    for i in time:
        t = i.text.strip()
        timing.append(t)
        
    # etracting maps links for coordinates
    link = box.find_all('a',class_ ="get_direction_click btn w-100 idx-info-card-str-dir-btn txt-color-3 bdr-color-6 bg-color-4 text-uppercase btn-hover-2")
    for i in link:
        ln = i.get('href')
        links.append(ln)
        
    for i in links:
        parsed_url = urllib.parse.urlparse(i)
        destination_param = urllib.parse.parse_qs(parsed_url.query)['destination'][0]
        lat, lng = destination_param.split(',')
        latitude.append(lat)
        longitude.append(lng)
        
# creating data frame
a = {"Store Name": store_name, "Address":address, "Phone Number":phone_no, "Timing":timing, "Latitude":latitude, "Longitude":longitude, "Links":links}
# removing error-Python Pandas ValueError Arrays Must be All Same Length
df1 = pd.DataFrame.from_dict(a, orient='index') 
    
```

```python
# adjusting the row and columns
df = df1.transpose()
```

```python
# keeping data till 66 rooms as there is no data except latitude/longitude
df2= df[:66]
```
